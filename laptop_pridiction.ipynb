```python
# applying feature engineering on screenresolution collumn
# making 3 diffrent collumns out of it 
# touch screen hai ya nahi ye dekh rhe
# resolutions ke liye 2 collumns bana rhe 
```

```python
df['Touchscreen'] = df['ScreenResolution'].apply(lambda x:1 if 'Touchscreen' in x else 0)
# lamda function se hamne naya collumn jiska nam touchscreen hai usme binary number se if else laga diya taki numeric data ho ske
```

```python
df.sample(5)
```

```python
df['Ips'] = df['ScreenResolution'].apply(lambda x:1 if 'IPS' in x else 0)
df['Full_HD'] = df['ScreenResolution'].apply(lambda x:1 if 'Full HD' in x else 0)
# same chiz ips aur HD check karne ke liye laga di
```

```python
new = df['ScreenResolution'].str.split('x',n=1,expand=True)
# yaha par ek nayi data frame banayi new naam ki (jo temporary hai)
# usme screenresolution vala collumn x ke basis par split kar diya 
# taki resolutions ko divide kar ske
# n=1 kiya hai taki X ki first aur bas ek hi occurence ke basis par devide ho(iss case mai farak nahi paad rha but 3 ressoltion hote toh padta)
```

```python
new.head(5)
```

```python
df['X_res'] = new[0]
df['Y_res'] = new[1]
```

```python
df.sample(5)
```

```python
df['X_res'] = df['X_res'].str.replace(',','').str.findall(r'(\d+\.?\d+)').apply(lambda x:x[0])
# ye karna pada taki x res vale collumn mai bas res res jaye full hd vagera sab haat jaye
# ye regular expression likhi hui hai jiska matlab hai ki sare numeric values jo dot ke pehle ho ya baad ho
# iske ek baar chatgpt kar lena
```

```python
df.head()
```

```python
df.drop(columns=['ScreenResolution'],inplace=True)
# ScreenResolution drop kar diya kyuki usko split karke sari imp info nikal li
```

```python
df['Full_HD'].value_counts()
```

```python
df['Ram'] = pd.to_numeric(df['Ram'].str.replace('GB', '', regex=False), errors='coerce').astype('Int32')
df['Weight'] = pd.to_numeric(df['Weight'].str.rstrip('kg'), errors='coerce').astype('float32')
# ram aur weight se kg aur gb hata diya taki datatype obj se int float kar ske
```

```python
df['X_res'] = df['X_res'].astype('int')
df['Y_res'] = df['Y_res'].astype('int')
```

```python
df.info()
```

```python
df['ppi'] = (((df['X_res']**2) + (df['Y_res']**2))**0.5/df['Inches']).astype('float')
# Pixels Per Inch ka formula laga diya
```

```python
df.drop(columns=['Inches','X_res','Y_res'],inplace=True)
# ppi nikal liya aab inn teeno ka kya kaam
```

```python
df['Cpu_Name'] = df['Cpu'].apply(lambda x:" ".join(x.split()[0:3]))
# isse top 3 words aa jayenge
# .join fun use kiya taki unn teeno words ko concatenate kar paye varna voh list ke alag alag element jaisa store hota
```

```python
df.head()
```

```python
df['Cpu_Name'].value_counts()
# isko dekh ke smjh aara hai ki i5,i3,i7,intel ka koi aur varna AMD  ke processor hai
```

```python
def fetch_processor(text):
    if text == 'Intel Core i7' or text == 'Intel Core i5' or text == 'Intel Core i3':
        return text
    else:
        if text.split()[0] == 'Intel':
            return 'Other Intel Processor'
        else:
            return 'AMD Processor'
```

```python
df['Cpu brand'] = df['Cpu_Name'].apply(fetch_processor)
```
